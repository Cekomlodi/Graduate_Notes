```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
#1
'a'
img = Image.open("AstroHW2_1a.jpeg")
fig,ax = plt.subplots(figsize = (30,30))
ax.imshow(img)
ax.set_ylim([1500,0])
'b'
r = 1 * 1.496e13 #cm
Mstar = 1 * 1.9884e33 #g
T = 100 #K
gasden = 1e-9 #g cm-3
s = 0.001 #cm
k = 1.3807e-16 #cm2 g s-2 K-1
mu = 2
mh =1.67e-24 #g
Vth = np.sqrt((8*k*T)/(np.pi*mu*mh))
soliden = 3 #g cm-3
G = 6.6743e-8 #cm3 g-1 s-2
Kepf = np.sqrt((G*Mstar)/(r**3))
Tsettle = (gasden*Vth)/(soliden*s*(Kepf**2))
#################
print('1)')
print('b. Tsettle = {0:.4f} yr'.format(Tsettle/3.154e+7))
print("disk lifetime is on the order of 10Myr before beginning to be considered MS stars")
print('This means that the time it takes the disk to settle vertically is << disk lifetime')
print('adding to the idea that solar system beginnings happen on very vast timescales ')
```

```python
#2
'a'
n = 3
cs = np.sqrt((k*T)/(mu*mh))
vk = np.sqrt((G*Mstar)/r)
nu = n*(cs/vk)**2
'b'
Vgas = vk*((1-nu)**0.5)
Vr_gas = Vgas - vk
'c'
size = np.array([.001, 20, 10000])
tfric = (soliden/gasden)*(size/Vth)
Tau_fric = tfric*Kepf
'd & f'
vr = ((Tau_fric**-1)*Vr_gas-nu*vk)/(Tau_fric+(Tau_fric**-1))
vr2 = (Vr_gas-nu*Tau_fric[0]*vk)
'e'
vr3 = (-nu*vk)/Tau_fric[2]
'g'
t_r = r/np.absolute(vr)
#################
print('2)')
print('a. nu = {0:.4f}'.format(nu))
print('b. Azimuthal Gas Velocity = {0:.4e} cm s-1'.format(Vgas))
print('Radial Gas Velocity = {0:.4e} cms-1'.format(Vr_gas))
print('c. Tau_friction for 10 micron = {0:.4e}'.format(Tau_fric[0]))
print('Tau_friction for 20 cm = {0:.4f}'.format(Tau_fric[1]))
print('Tau_friction for 100 m = {0:.4f}'.format(Tau_fric[2]))
print('d. V_r for 10 micron is << 1, full vr = {0:.4f}'.format(vr[0]))
print('post limit vr = {0:.4f}'.format(vr2))
print('e. V_r for 100 m is >> 1, full vr = {0:.4f}'.format(vr[2]))
print('post limit vr = {0:.4f}'.format(vr3))
print('f. The final V_r is 20 cm, full vr = {0:.4f}'.format(vr[1]))
print('g. Radial timescale for 10 micron = {0:.4f} years'.format(t_r[0]/3.154e+7))
print('Radial timescale for 20 cm = {0:.4f} years'.format(t_r[1]/3.154e+7))
print('Radial timescale for 100 m = {0:.4f} years'.format(t_r[2]/3.154e+7))
#Explain the significance of vr  Tau_fric^-1 for Tau_fric >>1
print("One reason vr  Tau_fric^-1 is significant for Tau_fric >>1 is because it mathematically shows that the velocity")
print("of an object as it gets bigger will decrease as more friction is applied. Therefore objects that have a larger size")
print("are more likely to fall inward.")
img2 = Image.open("AstroHW3written.jpeg")
fig,ax = plt.subplots(figsize = (30,30))
ax.set_ylim([1500,0])
ax.imshow(img2)
```

```python
#3
print('3)')
'a'
deltaV = 30 #cm s-1
r_p = 100000 #cm
rho_p = 3 #g cm-3
V = (4/3)*np.pi*r_p**3
M_p = rho_p*V

print('a. Inital Planetesimal Mass = {0:.4e} g'.format(M_p))
print('Inital Planetesimal Mass = {0:.4e} earth masses'.format(M_p/5.9722e27))

'b'
a = 5 * 1.496e13 #cm
#sig_g = 1 #g cm-2
sig_m = 1/100
R_b = (G*M_p)/(deltaV**2)
deltaM_p = np.pi*(R_b**2)*sig_m
t_b = R_b/deltaV
Tau_double = (M_p/deltaM_p)*t_b
factors = (M_p/deltaM_p)

print('b. Doubling timescale = {0:.4f} years'.format(Tau_double/3.154e+7))
print('b. Doubling timescale = {0:.4f} T_b'.format(factors))

'c'
M_t = (2.2e-3*(1**(3/2)*(1**3)))*5.972e+27 #g
counter = 0

while M_p < M_t:
    M_p_new = 2*M_p
    R_b_new = (G*M_p_new)/(deltaV**2)
    deltaM_P_new = np.pi*(R_b_new**2)*sig_m
    t_b_new = R_b_new/deltaV
    M_p = M_p_new
    R_b = R_b_new
    deltaM_p = deltaM_P_new
    t_b = t_b_new
    counter = counter+1
    
Tau_double_new = (M_p/deltaM_p)*t_b
print('c. Planetesimal Age = {0:.4f} years'.format((Tau_double_new/3.154e+7)*counter))

'd'
#M_p_dot = (3.8/5.9722e27*(((M_p/5.9722e27)/5.9722e27)**2/3)*(1**-1.5)*(1e6*3.154e+7)) #g s-1

M_earth = 5.9722e27 #g
AU = 5*1.496e13 #cm
g_s = M_earth/3.15576e+13 #g s-1
counter2 = 0
time = []
while M_p < 10*M_earth:
    M_p_new = 2*M_p
    R_b_new = (G*M_p_new)/(deltaV**2)
    M_P_dot_new = (3.8*g_s)*((M_p_new/M_earth)**(2/3))*((a/AU)**-1.5) #g s-1
    M_p = M_p_new
    R_b = R_b_new
    counter2 = counter2+1
    time.append((M_p/M_P_dot_new))

Tau_double_hill2 = np.sum(np.array(time))
#Tau_double_hill = (M_p/M_P_dot_new)
totalage = (Tau_double_hill)/3.15576e+13

print('d. Planetesimal Age for 10 earth masses = {0:.4f} Million years'.format((Tau_double_hill2)/3.15576e+13))
print('It is less about 3 Myr which means this planetesimal will graduate kindergarden and')
print('work towards being an adult planet :)')
```
