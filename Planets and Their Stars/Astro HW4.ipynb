```python
#Number 2
print('2)')
print('a.')
data = pd.read_csv('/Users/corinnekomlodi/Downloads/TNO.csv')
df = pd.DataFrame(data, columns = ['Prim. Desig.','a (AU)','e','i (deg)'])
PrimDesig, a, e, i = np.array(df['Prim. Desig.']), np.array(df['a (AU)']), np.array(df['e']), np.array(df['i (deg)'])

bins = np.arange(30,80,0.5)
fig, ax = plt.subplots(1, 1, figsize = (12,8))
ax.hist(a, bins = bins, color = 'silver')
ax.set_xlabel('a AU')
ax.set_ylabel('number of objects')

keplerPeriod = (a**3)**(1/2) #earth years
#p^2 = a^3

ax.axvline(x = 30, ymin = 0, ymax = 350, color = 'crimson', linestyle='--', label = '1:1 Trojans')
ax.axvline(x = ((2*165)**2)**(1/3), ymin = 0, ymax = 350, color = 'mediumturquoise', linestyle='--', label = '1:2 Plutinos')
ax.axvline(x = ((3*165)**2)**(1/3), ymin = 0, ymax = 350, color = 'pink', linestyle='--', label = '1:3 Plutinos')
ax.axvline(x = (((3/2)*165)**2)**(1/3), ymin = 0, ymax = 350, color = 'midnightblue', linestyle='--', label = '2:3 Plutinos')
ax.axvline(x = (((4/3)*165)**2)**(1/3), ymin = 0, ymax = 350, color = 'darkgreen', linestyle='--', label = '3:4')
ax.axvline(x = (((5/2)*165)**2)**(1/3), ymin = 0, ymax = 350, color = 'mediumpurple', linestyle='--', label = '2:5')

ax.legend()
print('roughly in order: 1:2, 1:1, 2:3, 1:3, 2:4, 2:5')
```

```python
print('b.')
AUrange = np.where((a >= 40)&(a <= 47))
fig, ax = plt.subplots(1, 1, figsize = (12,8))
ax.fill_between([0,4], [.2, .2], color = 'silver', alpha = .9, label = 'cold classical')
ax.fill_between([4,30], [.2, .2], color = 'teal', alpha = .8, label = 'hot classical')
ax.fill_between([30,max(i[AUrange])], [.2, .2], color = 'mediumaquamarine', alpha = .5, label = 'scattered disk')
ax.fill_between([0,max(i[AUrange])],[.2,.2], [1, 1], color = 'mediumaquamarine', alpha = .5)
ax.scatter(i[AUrange], e[AUrange], s = .5, color = 'k', label = 'KBO objects')
ax.set_xlabel("Inclination")
ax.set_ylabel('Eccentricity')
ax.set_xlim([0,max(i[AUrange])])
ax.set_ylim([0,1])
ax.legend(loc = 'center right')

wherecold = np.where((e[AUrange] <= 0.2)&(e[AUrange] >= 0)&(i[AUrange] <= 4)&(i[AUrange] >= 0))
wherehot = np.where((e[AUrange] <= 0.2)&(e[AUrange] >= 0)&(i[AUrange] <= 30)&(i[AUrange] >= 4))
wherescatter1 = np.where((e[AUrange] <= 0.2)&(e[AUrange] >= 0)&(i[AUrange] <= max(i[AUrange]))&(i[AUrange] > 30))
wherescatter2 = np.where((e[AUrange] >= 0.2)&(i[AUrange] <= max(i[AUrange]))&(i[AUrange] >= 0))
wherescatter = np.append(wherescatter1,wherescatter2)

percold = (len(e[wherecold])/len(e[AUrange]))*100
perhot = (len(e[wherehot])/len(e[AUrange]))*100
perscatter = (len(e[wherescatter])/len(e[AUrange]))*100
print('percentage cold = {0:.2f}%'.format(percold))
print('percentage hot = {0:.2f}%'.format(perhot))
print('percentage scatter = {0:.2f}%'.format(perscatter))
print(percold+perhot+perscatter)

#Overlay the following resonances:
#1:1 (Neptunes Trojans), 1:2, 1:3, 2:3 (Plutinos), 3:4, and 2:5. 
#Approximately order these six resonances from most to least populated.
```
